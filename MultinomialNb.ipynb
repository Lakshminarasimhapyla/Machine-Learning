```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import  MultinomialNB
from sklearn import metrics
import pandas as pd
```

```python
data=pd.read_csv("text_classification.csv",names=["message","label"])
```

```python
data
```

```python
data['labelnum']=data['label'].map({"pos":1,"neg":0})
```

```python
data
```

```python
inp_data=data['message']
output=data['labelnum']
```

```python
x_train,x_test,y_train,y_test=train_test_split(inp_data,output,test_size=0.3,random_state=5)
cv=CountVectorizer()
```

```python
x_train_1=cv.fit_transform(x_train)
x_test_1=cv.transform(x_test)
```

```python
df=pd.DataFrame(x_train_1.toarray(),columns=cv.get_feature_names_out())
```

```python
df
```

```python
clf=MultinomialNB()
```

```python
clf=clf.fit(x_train_1,y_train)
```

```python
pred=clf.predict(x_test_1)
```

```python
print("accuracy is :",metrics.accuracy_score(y_test,pred))
```

```python
print("confusion matrix :",metrics.confusion_matrix(y_test,pred))
```

```python
print("precesion is :",metrics.precision_score(y_test,pred))
```

```python
print("recall is :",metrics.recall_score(y_test,pred))
```
